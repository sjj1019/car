```python
# 死循环
import time

while True:
    print('hello itcast')
    
    time.sleep(1)
```

```python
# 通过变量控制循环
i = 0
while i < 100:
    print('i = %d' % i)
    i += 1
```

```python
# 判断，跳出循环，break

i = 0
while i < 100:
    print('i = %d' % i)
    
    if i > 10:
        break;
    
    i += 1
```

```python
# contiune

i = 0
while i < 100:
    
    i += 1
    
    if i > 10:
        continue;

    print('i = %d' % i)
    
print(i)
```

```python
# 嵌套
i = 0
while i < 5:
    j = 0
    while j < 3:
        print('j = %d' % j)
        j += 1
        
    i += 1
```
